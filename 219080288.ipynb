# Descriptive Stats Explained

```python
# Print definitions for statistical measures
print("1. Mean: It is the average value of a set of numbers.")
print("2. Median: It is the middle value in a set of numbers.")
print("3. Mode: It is the value that appears the most in a set of numbers.")
print("4. Weighted Mean: It is a type of average that gives different weights to different values in a set.")
print("5. Geometric Mean: It is a type of average that is calculated by multiplying all the numbers in a set and taking the root of the result.")
print("6. Range: It is the difference between the highest and lowest values in a set of numbers.")
print("7. Interquartile Range (IQR): It is the difference between Q3 and Q1 of a set of numbers.")
print("8. Variance: It is the measure of how spread out the values in a set are from the mean.")
print("9. Standard Deviation: It is the measure of how spread out the values in a set are from the mean, calculated as the square root of the variance.")
print("10. Coefficient of Variation: It is a measure of the spread of data.")
print("11. Covariance: It is a measure of the joint variability of two variables.")
print("12. Correlation coefficient: It is a measure of the strength and direction of the linear relationship between 2 variables.")
print("13. Percentiles: It describes how a score compares to other scores from the same number set.")
print("14. Detecting Outliers: It shows the values that are different from the other values.")
print("15. Quartiles: It is the values that divide a set of data into four equal parts.")
print("16. Distributed Shape: It shows the way data is spread out, such as symmetric, skewed.")
print("17. Histogram: It is a graph that shows the distribution of data by grouping it into intervals.")
print("18. Box Plot: It is a graph that shows the distribution of data using median, quartiles, and outliers.")
print("19. Five Number Summary: It shows the Minimum value, first quartile, median, third quartile, and the maximum value of a set of data.")
print("20. Z-Score: It is a standardized score that shows how many standard deviations a value is from the mean.")
print("21. Normal distribution: It shows the distribution of data using mean and standard deviation.")
print("22. Skewness: Measures the symmetry of the distribution. Positive skewness indicates a longer right tail, while negative skewness indicates a longer left tail.")
print("23. Kurtosis: Measures the 'tailedness' of the distribution. Positive kurtosis indicates heavier tails than a normal distribution, while negative kurtosis indicates lighter tails.")
print("24. Frequency Distribution: Displays the frequency of each value or range of values in a dataset.")
print("25. Probability Distribution Function (PDF): Defines the likelihood of a random variable taking a certain value or falling within a certain range.")
print("26. Cumulative Distribution Function (CDF): Gives the probability that a random variable is less than or equal to a certain value.")
print("27. Moments: Higher-order moments beyond the first two (mean and variance) can provide additional insights into the shape of the distribution.")
print("28. Confidence Intervals: Estimates the range within which a population parameter lies with a certain level of confidence.")
print("29. Coefficient of Determination (R-squared): Measures the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Normal distribution
plt.hist(df['Networth'], bins=30, density=True)
plt.title('Normal Distribution')
plt.xlabel('Networth')
plt.ylabel('Frequency')
plt.show()

# Skewness
skewness = stats.skew(df['Networth'])
print("Skewness:", skewness)

# Kurtosis
kurtosis = stats.kurtosis(df['Networth'])
print("Kurtosis:", kurtosis)

# Frequency Distribution
freq_dist, bins = np.histogram(df['Networth'], bins=10)
print("Frequency Distribution:", freq_dist)
print("Bins:", bins)

# Probability Distribution Function (PDF)
x = np.linspace(df['Networth'].min(), df['Networth'].max(), 1000)
pdf = stats.norm.pdf(x, loc=df['Networth'].mean(), scale=df['Networth'].std())
plt.plot(x, pdf)
plt.title('Probability Distribution Function (PDF)')
plt.xlabel('Networth')
plt.ylabel('Probability Density')
plt.show()

# Cumulative Distribution Function (CDF)
cdf = stats.norm.cdf(x, loc=df['Networth'].mean(), scale=df['Networth'].std())
plt.plot(x, cdf)
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Networth')
plt.ylabel('Cumulative Probability')
plt.show()

# Moments
moments = stats.moment(df['Networth'], moment=[1, 2, 3, 4])
print("First four moments:", moments)

# Confidence Intervals
confidence_interval = stats.norm.interval(0.95, loc=df['Networth'].mean(), scale=stats.sem(df['Networth']))
print("95% Confidence Interval:", confidence_interval)
```
